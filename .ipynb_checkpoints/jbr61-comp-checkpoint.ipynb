## PIZZA RESTAURANTS IN PITTSBURGH: OPEN AND CLOSED
    

In this notebook, I will try to find the best place to open a pizza shop in Pittsburgh, based off of the history of pizza shops in the area. My data set has all restruants in Pittsburgh. I was able to sort this down to just Pizza shops. This data set also shows the activity status of the restrautants. This will be really helpful for us, because it can show where this business has historically not done well.

I will decide the best place based off two things:
   * The ratio of closed to open restaurants
   * The number of active restaurants already in the area
So I will be looking for the neighborhood with the least amount of pizza places and lowest ratio of opened/closed pizza places.

### Closed Pizza Places

For this section the data will be rperesented on a scale from 1 to 7, as that is how the data set marks them as open or closed. I will find the average of the status of all shops in the neighborhoods. The neighborhood with the numbers closest to 7 will have the most closed shops and close to 1 means very little closed shops.

```python
import pandas as pd
import numpy as np
import geopandas
%matplotlib inline
import matplotlib.pyplot as plt

neighborhoods_map = geopandas.read_file("Neighborhoods/Neighborhoods_.shp")
closed_pizza = pd.read_csv("Sorted_DataSheet.csv")
open_pizza = pd.read_csv("SortedBook2.csv")

```

```python
hoods_ratio = closed_pizza.groupby("municipal").mean()['status']
hoods_ratio_map = neighborhoods_map.merge(hoods_ratio, how='left', left_on='hood', right_on='municipal')
hoods_ratio_map[['hood','status']].sort_values('status', ascending=True).head(7)
```

From this table we are able to tell that the neighborhood with the most pizza shop success is East Liberty.

```python
hoods_ratio_map.plot(column='status',
               cmap='YlGn',
               edgecolor="white",
               legend=True,
               legend_kwds={'label': "Likelihood of Closed Pizza Shop"},
               figsize=(15, 10),
               missing_kwds={"color": "black"}
               )
```

This map is also very helpful in out search. It shows that shops in the downtown area have a much higher turnover rate. This is likely due to high rent and running costs in that area. The map also shows it get lighter in the areas around downtown, possibly because these areas still get a lot of traffic from downotown, but nor the same operation costs. Finally, it seems to get dark again on the outskirts of city limits. So we also might want to look for an area closer to downtown, but not in it.

### Pizza Competitors

In this section I will try to find the neighborhood with the least amount of competition. More pizza shops means more demand, but less market share.

```python
list_of_un_mun = open_pizza['municipal'].unique()
```

```python
mun_dict = dict()

for mun in list_of_un_mun:
    count = open_pizza['municipal'].value_counts()[mun]
    mun_dict[mun] = count
```

```python
keys = mun_dict.keys()
values = mun_dict.values()

data = dict()

data['hood'] = keys
data['open'] = values
```

```python
mun_df = pd.DataFrame.from_dict(data)
```

```python
hoods_open = open_pizza.groupby("municipal").count()
hoods_open_map = neighborhoods_map.merge(mun_df, how='left', left_on='hood', right_on='hood')
hoods_open_map[['hood','open']].sort_values('open', ascending=False).head()
```

This tells us the worst places to start a business.

```python
hoods_open_map.plot(column='open',
               cmap='YlGn',
               edgecolor="white",
               legend=True,
               legend_kwds={'label': "Number of Open Pizza Shops"},
               figsize=(15, 10),
               missing_kwds={"color": "black"}
               )
```

We would want to target areas that are lighter in color, if we were going by just this metric. However, we must compare it to our other metrics!
